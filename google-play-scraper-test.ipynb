```python
# from google_play_scraper import Sort, reviews
# from urllib.parse import urlparse, parse_qs
# import pandas as pd
# import re

# isRecent = True

# STARTDATE = pd.Timestamp('2024-08-01T00:00:00')
# ENDDATE = pd.Timestamp('2024-09-01T00:00:00')
# reviewDf = pd.DataFrame()
# token = None

# # 크롤링하고자 하는 어플리케이션 목록 엑셀 파일
# file_name = 'application_list.xlsx'
# application_list = pd.read_excel(file_name)

# APPNAMES = list(application_list['어플'])
# URLS = list(application_list['링크'])

# for URL, NAME in zip(URLS, APPNAMES) :

#     print(f"{NAME} 스크롤 시작")

#     # URL 파싱
#     parsed_url = urlparse(URL)
#     # 쿼리 파라미터 추출
#     query_params = parse_qs(parsed_url.query)
#     # id 값 추출
#     app_id = query_params.get('id', [None])[0]

#     while isRecent :
#         # google_play_scraper 리뷰 한개씩 추출
#         result, continuation_token = reviews(
#             app_id,
#             lang='ko',
#             country='kr',
#             sort=Sort.NEWEST,
#             count=1,
#             continuation_token=token # 리뷰 크롤링을 이어가기 위한 토큰
#         )
#         token = continuation_token

#         df = pd.DataFrame(result)
#         review_date = df.iloc[0]['at']

#         if STARTDATE <= review_date :
#             if ENDDATE >= review_date :
#                 reviewDf = pd.concat([reviewDf, df])
#         else :
#             isRecent = False

#     # 인덱스 재배열
#     reviewDf.reset_index(inplace=True, drop=True)
#     # 파일 이름에 사용할 수 없는 특수문자 제거
#     filename = re.sub('[\/:*?"<>|]','',NAME)
#     reviewDf.to_excel(f'review_{filename}.xlsx')

from google_play_scraper import app
from urllib.parse import urlparse, parse_qs
import pandas as pd
import re

# 엑셀 파일에서 어플리케이션 목록 읽기
file_name = 'application_list.xlsx'
application_list = pd.read_excel(file_name)

# 엑셀 파일의 '어플' 열에 있는 앱 이름과 '링크' 열에 있는 앱 경로 추출
APPNAMES = list(application_list['어플'])
URLS = list(application_list['링크'])

# 결과 저장할 리스트
app_details = []

for URL, NAME in zip(URLS, APPNAMES):

    print(f"{NAME} 정보 가져오는 중...")

    # URL 파싱
    parsed_url = urlparse(URL)
    # 쿼리 파라미터 추출
    query_params = parse_qs(parsed_url.query)
    # id 값 추출
    app_id = query_params.get('id', [None])[0]

    if app_id:
        # google_play_scraper에서 앱 디테일 정보 추출
        result = app(
            app_id,
            lang='ko',  # 언어 설정
            country='kr'  # 국가 설정
        )
        
        # 필요한 정보만 추출
        app_info = {
            'App Name': result['title'],
            'Developer': result['developer'],
            'Description': result['description'],
            'Installs': result['installs'],
            'Ratings': result['ratings'],
            'Reviews': result['reviews'],
            'Score': result['score'],
            'Version': result['version'],
            'Updated': pd.to_datetime(result['updated'], unit='s'),  # 타임스탬프를 날짜로 변환
            'App ID': result['appId'],
            'Genre': result['genre']
        }

        # 추출한 정보를 리스트에 추가
        app_details.append(app_info)

    else:
        print(f"앱 ID를 찾을 수 없습니다: {NAME}")

# 데이터프레임으로 변환
app_df = pd.DataFrame(app_details)

# 데이터프레임 확인
print(app_df)

# 파일 이름에 사용할 수 없는 특수문자 제거
filename = re.sub('[\/:*?"<>|]', '', 'app_details')

# 결과를 엑셀 파일로 저장
app_df.to_excel(f'{filename}.xlsx', index=False)
```
